### However the model cant accept the data in this format so we need to restructure the data into a matrix of features where each column is training point of n_features.

So, what we have to do is to flatten the array and this is a common way to do it.

```python
x = train_x_raw.reshape((train_x_raw.shape[0],-1)).T
x.shape
```

now we have a matrix in the correct format for the neural network to work with, anyways this part is implemented in the load_data script and we can directly load the data.

```python
train_x,train_y,test_x,test_y,classes = load_data.load_data()
```

```python
print('shape of training features: ',train_x.shape)
print('shape of training labels: ',train_y.shape)
print('shape of testing features: ',test_x.shape)
print('shape of testing lables: ',test_y.shape)
```

## Running the neural network

1- we choose the dimentions of each layer and pass them as a list ,each element = number of neurons for layer, i chose a 5-layer neural network<br>

#### Ofcourse the input layer must match the dimention of the features 12288

```python
# layers dims
layers_dims = [12288 , 15, 12, 7, 5, 2, 1]
layers_dims
```

2- intialize the neural network object with the training set and layers dimentions, we divide by 255 to normalize the training set (pixel values between 0-255).

```python
network = DNN(layers_dims,X_train=train_x/255,y_train=train_y)
```

3- run the training method for the neural network.<br><br>
parameters are :<br>

<ul>
    <li>learning rate = 0.0095</li>
    <li>number of iterations = 3000</li>
    <li>print_cost = True</li>

```python
network.train_network(lr = 0.0095,n_iters = 3000,print_cost = True)
```

4- now we can plot the costs vs iterations by accessing the values of costs and iterations in the network object<br><br>
note : we can plot the costs vs iters directly by setting plot_cost = True

```python
fig,ax = plt.subplots(figsize = (10,5))
plt.plot(network.iters,network.costs)
plt.xlabel('number of iterations')
plt.ylabel('cost')
plt.title('costs vs iterations');
```

Great! the neural network is minimizing the cost and working fine!<br><br>
now we can make predictions and get scores.<br><br>

5- make predictions by using the predict method on the network object

```python
# predictions on test set
pred_test = network.predict(test_x/255)
# predictions on train set
pred_train = network.predict(train_x/255)

pred_test
```

6- get the accuracy score for training and testing using accuracy method on the network object

```python
acc_test = network.accuracy(test_y,pred_test)
acc_train = network.accuracy(train_y,pred_train)

print('test acc : ',acc_test)
print('train acc : ',acc_train)
```

#### we got a 96% accuracy on the training set and 82% accuracy on the testing set.
Which is not bad for a non regularized model !
